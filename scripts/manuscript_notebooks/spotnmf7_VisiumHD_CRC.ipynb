```python
import seaborn as sns
import matplotlib.pyplot as plt

def plot_cell_type_percentage_per_patient(df, cell_type):
    # Total counts per patient
    total_counts = df['Patient'].value_counts().rename('Total')

    # Counts of the specific cell type per patient
    cell_counts = df[df['Level2'] == cell_type]['Patient'].value_counts().rename('CellType')

    # Combine into one DataFrame
    percent_df = pd.concat([total_counts, cell_counts], axis=1).fillna(0)
    percent_df['Percentage'] = 100 * percent_df['CellType'] / percent_df['Total']
    percent_df = percent_df.reset_index().rename(columns={'index': 'Patient'})

    # Sort by patient name
    percent_df = percent_df.sort_values('Patient')

    # Plot
    sns.barplot(data=percent_df, x='Patient', y='Percentage', palette='coolwarm')
    plt.title(f'Percentage of "{cell_type}" Cells Within Each Patient')
    plt.xticks(rotation=45)
    plt.ylabel('Percentage (%)')
    plt.tight_layout()
    plt.show()

plot_cell_type_percentage_per_patient(metadata, "CD8 T cell")
```

```python
import os 
import pandas as pd 
import scanpy as sc 
import sys
import numpy as np
sys.path.append('..')
from Wassgard.wassgard import  io, pl, annotate
import ast
import seaborn as sns
import matplotlib.pyplot as plt

results_dir = r"Z:\MorrissyLab Dropbox\Visium_profiling\crc_visiumhd_case_study\P5_CRC_results_lr0.01"
output_dir = r"Z:\MorrissyLab Dropbox\Manuscripts_in_progress\spOT-NMF\Raw_Data\visiumhd_casestudy_fig8"


# Prepare a list to collect all data
all_data = []

for sample_dir in os.listdir(results_dir):
    genesets_file = os.path.join(results_dir, sample_dir, f"genesets_results/ranked_genesets_{sample_dir}_human_crc_level2.csv")
    
    if not os.path.exists(genesets_file):
        continue

    df = pd.read_csv(genesets_file, index_col=0)
    
    try:
        top_marker_genes = df['0'].apply(lambda x: ast.literal_eval(x)[1]).values
        for gene in top_marker_genes:
            all_data.append({"rank": sample_dir.replace("P5_CRC_GRCh38_","").replace("_24um","").replace("CRC_ALL_GRCh38_",""), "scores": gene})
    except Exception as e:
        print(f"Skipping {sample_dir} due to error: {e}")

# Convert to DataFrame
plot_df = pd.DataFrame(all_data)

# Create plots
plt.figure(figsize=(12, 6))
sns.violinplot(data=plot_df, x="rank", y="scores", inner=None, color='skyblue')
sns.stripplot(data=plot_df, x="rank", y="scores", color='black', alpha=0.6, jitter=True)

plt.xticks(rotation=45)
plt.title("Top Marker Genes scores per rank lr")
plt.ylabel("Score")
plt.tight_layout()
plt.show()

threshold = 0.2
counts_above_threshold = plot_df[plot_df['scores'] > threshold].groupby('rank').size()

# Print result
print(counts_above_threshold)

import matplotlib.pyplot as plt
import numpy as np

# Define thresholds
thresholds = np.arange(0.0, 1.01, 0.05)

# Initialize plot
plt.figure(figsize=(10, 6))

# Loop through ranks
for rank in sorted(plot_df['rank'].unique()):
    rank_df = plot_df[plot_df['rank'] == rank]
    counts = []

    for thresh in thresholds:
        count = (rank_df['scores'] > thresh).sum()
        counts.append(count)

    plt.plot(thresholds, counts, label=f"Rank {rank}")

# Plot formatting
# plt.axvline(0.3, color='gray', linestyle='--', label='Threshold = 0.3')
plt.xlabel("Score Threshold")
plt.ylabel("Count of Scores > Threshold")
plt.title("Score Threshold Sweep per Rank")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Figure 8 c) Computational 

```python
import os
import re
import pandas as pd
import matplotlib.pyplot as plt
from hurry.filesize import size

# Directory containing the HVG results
results_dir = r"Z:\MorrissyLab Dropbox\Visium_profiling\crc_visiumhd_case_study\P5_CRC_results_lr0.001"
# Prepare regex to parse lines
pattern = re.compile(
    r"\|.*\|\s*(?P<iter>\d+)/(?:\d+) \[(?P<time>[\d:]+)<.*loss=(?P<loss>[0-9\.e+-]+),.*gpu_memory_allocated=(?P<gpu>[0-9\.e+-]+)\]"
)

# Container for data per sample
data = {}

# Parse logs
for sample in os.listdir(results_dir):
    sample_dir = os.path.join(results_dir, sample)
    log_file = os.path.join(sample_dir, "logs", "wassgard.err")
    if not os.path.isfile(log_file):
        continue

    records = []
    # Open with encoding fallback
    with open(log_file, 'r', encoding='utf-8', errors='ignore') as f:
        for line in f:
            m = pattern.search(line)
            if m:
                it = int(m.group('iter'))
                # h, m_, s = map(int, m.group('time').split(':'))
                # elapsed = h*3600 + m_*60 + s
                elapsed = m.group('time')
                loss = float(m.group('loss'))
                gpu = float(m.group('gpu'))
                records.append((it, elapsed, loss, gpu))

    if records:
        df = pd.DataFrame(records, columns=['iteration', 'elapsed_time', 'loss', 'gpu_memory_allocated'])
        data[sample] = df

# Plot and save each sample's metrics in its own folder
for sample, df in data.items():
    sample_dir = os.path.join(results_dir, sample)
    os.makedirs(sample_dir, exist_ok=True)
    fig, ax1 = plt.subplots(figsize=(8, 4))
    ax1.plot(df['iteration'], df['loss'], label='Loss')
    ax1.set_xlabel('Iteration')
    ax1.set_ylabel('Loss', color='tab:blue')

    ax2 = ax1.twinx()
    ax2.plot(df['iteration'], df['gpu_memory_allocated'], label='GPU Mem', linestyle='--')
    ax2.set_ylabel('GPU Memory Allocated (bytes)', color='tab:orange')

    plt.title(f"Sample: {sample}")
    fig.tight_layout()

    # Save as PDF in sample folder
    out_path = os.path.join(sample_dir, f'training_metrics_plot_{sample}.pdf')
    fig.savefig(out_path)
    plt.close(fig)
    # print(f"Saved plot for {sample} to {out_path}")

# Build and print summary DataFrame of last iteration per sample
summary_records = []
for sample, df in data.items():
    last_row = df.loc[df['iteration'].idxmax()]
    summary_records.append({
        'sample': sample,
        'last_iteration': int(last_row['iteration']),
        'elapsed_time': last_row['elapsed_time'],
        'loss': float(last_row['loss']),
        'gpu_memory_allocated': size(float(last_row['gpu_memory_allocated']))
    })

summary_df = pd.DataFrame(summary_records)
summary_df
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# --- Sample Preprocessing ---
summary_df['rank'] = summary_df['sample'].str.extract(r'_([0-9]+)_')[0].astype(int)
summary_df['bin_size'] = summary_df['sample'].str.extract(r'_([0-9]+um)')[0]
summary_df['bin_size'] = pd.Categorical(summary_df['bin_size'], categories=['8um', '16um', '24um'], ordered=True)

def hms_to_seconds(hms_str):
    h, m, s = map(int, hms_str.split(':'))
    return h * 3600 + m * 60 + s

summary_df['elapsed_seconds'] = summary_df['elapsed_time'].apply(hms_to_seconds)
summary_df['time_per_iter'] = summary_df['elapsed_seconds'] / summary_df['last_iteration']

summary_df['gpu_memory_allocated'] = pd.Categorical(summary_df['gpu_memory_allocated'],
                                                    categories=['12G', '11G', '7G'], ordered=True)

# --- Plotting ---
sns.set_theme(style="whitegrid", font_scale=1.4)
palette = sns.color_palette("colorblind")

fig, axes = plt.subplots(1, 3, figsize=(10.5, 3.5), constrained_layout=True)
metrics = ['loss', 'time_per_iter', 'gpu_memory_allocated']
titles = ['Final Loss', 'Time per Iteration (s)', 'GPU Memory Allocation']

for ax, metric, title in zip(axes, metrics, titles):
    sns.boxplot(data=summary_df, x='bin_size', y=metric, ax=ax,
                color='lightgray', width=0.6, fliersize=0, linewidth=1)
    
    sns.stripplot(data=summary_df, x='bin_size', y=metric, ax=ax,
                  jitter=True, marker='o', alpha=0.8, color='black',
                  size=4.5, linewidth=0.4, edgecolor='gray')

    ax.set_title(title, fontsize=14, weight='bold')
    ax.set_xlabel("Bin Size", fontsize=12)
    ax.tick_params(axis='x', rotation=45)
    ax.set_ylabel(title if metric == 'loss' else "", fontsize=12)

# Suptitle
fig.suptitle("Effect of Bin Size on Final Loss, Iteration Time, and GPU Memory", fontsize=16, weight='bold', y=1.05)

# Save and show
fig.savefig(os.path.join(figure_dir, "memory_loss_time_bin_boxplot.pdf"), bbox_inches="tight", dpi=300)
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os

# Ensure bin_size order
ordered_bins = ['8um', '16um', '24um']
plot_df['bin_size'] = pd.Categorical(plot_df['bin_size'], categories=ordered_bins, ordered=True)

# Compute fractions above thresholds
thresholds = np.arange(0.05, 0.41, 0.05)
records = []
for (bin_size, rank), group in plot_df.groupby(['bin_size', 'rank']):
    total = len(group)
    for th in thresholds:
        count_above = (group['scores'] > th).sum()
        frac_above = count_above / total
        records.append({
            'bin_size': bin_size,
            'rank': int(rank),
            'threshold': th,
            'fraction_above': frac_above
        })

threshold_df = pd.DataFrame(records)

# Choose threshold to display
th = 0.1
thresh_df = threshold_df[threshold_df['threshold'] == th]

# Pivot: now rows = rank, columns = bin_size (rotated)
heatmap_data = thresh_df.pivot(index='rank', columns='bin_size', values='fraction_above')
heatmap_data = heatmap_data[ordered_bins]  # Ensure correct column order

# Plot
plt.figure(figsize=(5, 3.5))
sns.set(style='whitegrid', font_scale=1.2)
ax = sns.heatmap(
    heatmap_data,
    annot=True,
    fmt=".2f",
    cmap='YlGnBu',
    cbar_kws={'label': f'Fraction > {th:.2f}'},
    linewidths=0.5,
    linecolor='gray'
)

# Titles and labels
plt.title(f'Proportion of Gene-Sets with Scores > {th}', fontsize=14, fontweight='bold')
plt.xlabel('Bin Size', fontsize=12)
plt.ylabel('Rank', fontsize=12)
plt.xticks(rotation=0)
plt.yticks(rotation=0)

plt.tight_layout()
plt.savefig(os.path.join(figure_dir, f"genesets_bins_ranks_rotated={th}.pdf"), bbox_inches="tight", dpi=300)
plt.show()
```

```python
import os 
import pandas as pd 
import scanpy as sc 
import sys
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import math

sys.path.append('..')
from Wassgard.wassgard import io, pl, annotate

results_dir = r"Z:\MorrissyLab Dropbox\Visium_profiling\crc_visiumhd_case_study\P5_CRC_results_lr0.001"

for sample_dir in os.listdir(results_dir):
    output_path = io.check_dir(os.path.join(results_dir, sample_dir, "density_plots_v2"))
    usage_file = os.path.join(results_dir, sample_dir, f"topics_per_spot_{sample_dir}.csv")
    if not os.path.exists(usage_file):
        print(f"File not found: {usage_file}")
        continue

    df = pd.read_csv(usage_file, index_col=0)
    num_cols = df.shape[1]
    num_rows = math.ceil(num_cols / 5)

    ### Version 1: Only values above threshold
    fig1, axes1 = plt.subplots(nrows=num_rows, ncols=5, figsize=(15, 3 * num_rows), squeeze=False)
    axes1 = axes1.flatten()

    ### Version 2: All values with threshold line
    fig2, axes2 = plt.subplots(nrows=num_rows, ncols=5, figsize=(15, 3 * num_rows), squeeze=False)
    axes2 = axes2.flatten()

    ### Version 3: All values with threshold, zoomed in (0–0.4)
    fig3, axes3 = plt.subplots(nrows=num_rows, ncols=5, figsize=(15, 3 * num_rows), squeeze=False)
    axes3 = axes3.flatten()

    for i, col in enumerate(df.columns):
        threshold = df[col].quantile(0.9)
        filtered_values = df[col][df[col] > threshold]

        # Plot 1: Only values above 90th percentile
        if filtered_values.empty:
            axes1[i].text(0.5, 0.5, f'No data > 90th percentile', ha='center', va='center')
            axes1[i].set_xlim(0, 1)
        else:
            sns.histplot(filtered_values, kde=True, ax=axes1[i], bins=50, color='skyblue', edgecolor='black')
            axes1[i].set_xlim(0, 1)
        axes1[i].set_title(col)
        axes1[i].set_xlabel("Usage > 90th")
        axes1[i].set_ylabel("Density")

        # Plot 2: All values, threshold line
        sns.histplot(df[col], kde=True, ax=axes2[i], bins=50, color='lightgray', edgecolor='black')
        axes2[i].axvline(threshold, linestyle='--', color='red', label='90th percentile')
        axes2[i].set_xlim(0, 1)
        axes2[i].set_title(col)
        axes2[i].set_xlabel("Usage")
        axes2[i].set_ylabel("Density")

        # Plot 3: Same as plot 2 but zoomed into [0, 0.4]
        sns.histplot(df[col], kde=True, ax=axes3[i], bins=50, color='lightgray', edgecolor='black')
        axes3[i].axvline(threshold, linestyle='--', color='red', label='90th percentile')
        axes3[i].set_xlim(0, 0.4)
        axes3[i].set_title(col)
        axes3[i].set_xlabel("Usage")
        axes3[i].set_ylabel("Density")

    # Remove unused subplots for each figure
    for j in range(i + 1, len(axes1)):
        fig1.delaxes(axes1[j])
        fig2.delaxes(axes2[j])
        fig3.delaxes(axes3[j])

    # Titles and save
    fig1.suptitle(f"Only Values > 90th Percentile - {sample_dir}", fontsize=16)
    fig2.suptitle(f"All Values with Threshold - {sample_dir}", fontsize=16)
    fig3.suptitle(f"Zoomed (0–0.4) with Threshold - {sample_dir}", fontsize=16)

    plt.tight_layout(rect=[0, 0, 1, 0.96])
    fig1.savefig(os.path.join(output_path, f"usage_density_{sample_dir}_only_th90.pdf"), dpi=300)
    fig2.savefig(os.path.join(output_path, f"usage_density_{sample_dir}_with_th90.pdf"), dpi=300)
    fig3.savefig(os.path.join(output_path, f"usage_density_{sample_dir}_with_th90_zoomed.pdf"), dpi=300)

    plt.close(fig1)
    plt.close(fig2)
    plt.close(fig3)
```
